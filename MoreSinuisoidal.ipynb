```
p= np.pi

x = np.arange(-2*p, 2*p, .01)
y = [5*np.sin(i) for i in x]
a = [np.sin(i+5) for i in x]
b = [np.sin(.5*i) for i in x]
c = [np.sin(i) + 5 for i in x]
d = [5*np.sin(.5*i-4) -3 for i in x]
e = [np.sin(9*i) for i in x]
```

```
plt.figure(figsize=(15,10))
plt.plot(x,y,label='y = 5sin(x)')
plt.plot(x,a, label='y = sin(x+5)')
plt.plot(x,b, label='y = sin(1/2x)')
plt.plot(x,c, label='y = sin(x) + 5')
plt.plot(x,d, label='y = 5sin(1/2x-4)-3')
plt.plot(x,e, label='y = sin(9x)')
plt.title('Graph of More Sine Functions')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```
x = np.arange(-2*p, 2*p, .01)
y = [np.sin(i**2) for i in x]
a = [(np.sin(i))**2 for i in x]
```

```
plt.figure(figsize=(15,10))
plt.plot(x,y,label='y = sin(x**2)')
plt.plot(x,a, label='y = (sin(x))**2')

plt.title('Graph of More Sine Functions')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```
